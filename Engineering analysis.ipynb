```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Engineering.csv")
df['bookprice']=df['bookprice'].replace({'â‚¹':'',',':''},regex=True).astype(float)
```

```python
plt.figure(figsize=(11,7))
plt.hist(df['bookprice'])
sns.countplot(data=df, x='bookprice',palette='viridis')
plt.xticks(rotation=90)
plt.title('Frequency of Each Book Price')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
top_expensive_books = df.nlargest(10, 'bookprice')

plt.figure(figsize=(12, 8))
sns.barplot(data=top_expensive_books, x='bookprice', y='Bookname', palette='coolwarm')
plt.title('Top 10 Most Expensive Books')
plt.xlabel('Price')
plt.ylabel('Title')
plt.show()
```

```python
top_cheapest_books = df.nsmallest(10, 'bookprice')
plt.figure(figsize=(11,8))
sns.barplot(data= top_cheapest_books, x='bookprice',y='Bookname',palette='coolwarm')
plt.title('Top 10 Most cheapest Books in Comics and Manga')
plt.xlabel('bookprice')
plt.ylabel('Bookname')
plt.show()
```
